```python
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd

# === Open the monthly max temperature data for QLD ===
ds = xr.open_dataset(r'Data\Farm Data\3_Cleaned_Tas_Farm_Data\Aus_farm_data_1991_cropped.nc')
farmland = ds['farmland_per_cell']  # Adjust variable name as needed

plt.figure(figsize=(10, 6))
farmland.plot(cmap='Greens')
plt.title('Farmland per Cell - 1991')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# === Load the shapefile for Tasmania ===
shapefile_path = r'Data\Tas_Shapefile\tasmania.shp'
tas_shape = gpd.read_file(shapefile_path).to_crs("EPSG:4326")

# === Open the cropped farm dataset ===
ds = xr.open_dataset(r'Data\Farm Data\3_Cleaned_Tas_Farm_Data\Aus_farm_data_1991_cropped.nc')
farmland = ds['farmland_per_cell']

# === Create figure and axis ===
fig, ax = plt.subplots(figsize=(10, 6))

# === Plot the farmland raster first ===
farmland.plot(ax=ax, cmap='viridis')

# === Overlay Tasmania boundary ===
tas_shape.boundary.plot(ax=ax, color='black', linewidth=0.8, alpha=0.8)

# === Labels ===
ax.set_title('Farmland per Cell - 1991')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
